```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
```

```python
def loadResultData(savedPath) :
    #Load Logistic Regression Results dataset
    with open(savedPath, "rb") as input_file:
        dataDict = pickle.load(input_file)

    return dataDict["steps"], dataDict["lossVals"], dataDict["accVals"], dataDict["accTest"]

#Load Logistic Regression Results dataset
steps, simpleLgLoss, simpleLgAcc, simpleLgTestAcc = loadResultData("../data/logisticRegressionModelMetrics.pickle")
print("Simple Logistic regression Model metrics loaded")

#Load Ridge Logistic Regression Results dataset
_, ridgeLgLoss, ridgeLgAcc, ridgeLgTestAcc = loadResultData("../data/ridgeRegressionModelMetrics.pickle")
print("Ridge Logistic regression Model metrics loaded")

#Load Lasso Logistic Regression Results dataset
_, lassoLgLoss, lassoLgAcc, lassoLgTestAcc = loadResultData("../data/lassoRegressionModelMetrics.pickle")
print("Lasso Logistic regression Model metrics loaded")

#Load Elastic Net Logistic Regression Results dataset
_, elasticLgLoss, elasticLgAcc, elasticLgTestAcc = loadResultData("../data/elasticNetRegressionModelMetrics.pickle")
print("Elastic Net Logistic regression Model metrics loaded")
```

```python
fig = plt.figure(figsize=(15, 6), dpi=80)
lossPlt = fig.add_subplot(1, 2, 1)
accPlt = fig.add_subplot(1, 2, 2)

lossPlt.plot(steps, simpleLgLoss, label="Simple Regression")
lossPlt.plot(steps, ridgeLgLoss, label="Ridge Regression")
lossPlt.plot(steps, lassoLgLoss, label="Lasso Regression")
lossPlt.plot(steps, elasticLgLoss, label="Elastic Net Regression")
lossPlt.set_xlabel('Steps')
lossPlt.set_ylabel('Loss')
lossPlt.set_title('Loss Analysis')
lossPlt.legend()

accPlt.plot(steps, simpleLgAcc, label="Simple Regression")
accPlt.plot(steps, ridgeLgAcc, label="Ridge Regression")
accPlt.plot(steps, lassoLgAcc, label="Lasso Regression")
accPlt.plot(steps, elasticLgAcc, label="Elastic Net Regression")
accPlt.set_xlabel('Steps')
accPlt.set_ylabel('Accuracy')
accPlt.set_title('Accuracy Analysis')
accPlt.legend()

plt.show()
```

```python
modelTypes = ('Simple', 'Lasso', 'Ridge', 'ElasticNet')
y_pos = np.arange(len(modelTypes))
performance = [simpleLgTestAcc, lassoLgTestAcc, ridgeLgTestAcc, elasticLgTestAcc]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, modelTypes)
plt.ylabel('Accuracy')
plt.title('Accuracy on Test dataset')

plt.show()
```
